```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("//kaggle//input//titanic//train.csv")
df.head()
gc=df.groupby(['Sex','Survived']).size().unstack()
gc.columns=['not survived','survived']
gc
gc.plot(kind='bar')
plt.xlabel('gender')
plt.ylabel('Number of passengers')
plt.title('graph of survival based on gender')
plt.show()
df.describe()
bins=[0,12,25,50,100]
labels=['child','teen','youth','elder']
df['agegrp']=pd.cut(df['Age'],bins=bins,labels=labels)
ac=df.groupby(['agegrp','Survived']).size().unstack()
ac
cc=df.groupby(['Pclass','Survived']).size().unstack()
cc.index=('1st class','2nd class','3rd class')
cc
cc.plot(kind='bar')
plt.xlabel('passenger class')
plt.ylabel('number of surivers')
plt.title('number of people survived in each class')
plt.xticks([0,1,2],['1ST CLASS','2ND CLASS','3RD CLASS'],rotation=0)
plt.show()

```
